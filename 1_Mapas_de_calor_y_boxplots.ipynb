```
import pandas as pd
# GRAFICAS
from matplotlib import pyplot as plt
import seaborn as sb
```

```
df = pd.read_excel('data.xlsx')
```

```
# HISTOGRAMA
%matplotlib inline 
df[['Season ', 'Episode ']].plot.hist(bins=9, alpha=0.5)
plt.show
# Esta grafica nos menciona cuantos episodios 9, 8 7, hay en general 
# toda la serie y la cantidad de temporadas actuales (5)
```

```
plt.plot(df.IMDB, df['Watch Time (minutes)'], 'o') # en x va pago y en y gastos, graficando con *
plt.xlabel('Ranking')
plt.ylabel('Duracion')
plt.show()
# Ejemplo de interpretación: cuando dan un puntaje de 7, el episodio duro alrededor de 45 min
```

```
# dada la grafica anterior podemos darnos cuenta que no se cuenta con gran correlacion
df.corr(method='pearson')
```

```
# BIGOTES PARA IMDB 
sb.set_theme(style='darkgrid') # diseño
IMDB = df.IMDB
ax = sb.boxplot(x=df.IMDB)
# Obtenemos una mediana es 8.3 
# Hay pocos ranking que se encuentran en el tercer cuartil, mientras que la mayoria se encuentra en el 4to
# en el primer y segundo cuartil esta aproximandamente la misma cantidad
```

```
# BIGOTES PARA WATCH TIME 
sb.set_theme(style='darkgrid') # diseño
watch_time = df['Watch Time (minutes)']
ax = sb.boxplot(x=df['Watch Time (minutes)'])
# Obtenemos una mediana es 47.5 
# Pocos capitulos van de 40-42.5 minutos de duracion, en 42.5-52.5 de igual manera (cuartil 2 y 3)
# la mayoria de los capitulos es de 52 a 60 min
```

```
# BIGOTES PARA WATCH TIME 
sb.set_theme(style='darkgrid') # diseño
episode = df['Episode ']
ax = sb.boxplot(x=df['Episode '])
# Obtenemos una mediana es 47.5 
# Hay varias temporadas que tienen entre 6-9 capitulos, nulas entere 0-2 y
# Una distribucion uniforme entre el cuartil dos y tres
```

```
# MAPA DE COLOR
colormap = plt.cm.viridis # esquema de color
plt.figure(figsize=(12,12))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sb.heatmap(df.corr(method='pearson'), linewidths=0.1, vmax=1.0, square=True, 
           cmap=colormap, linecolor='white', annot=True);
# 1.- La unica correlacion que se aprecia es con año y season, en donde vemos que hay correlacion
# con el tiempo en sacar una nueva temporada por año
# 2.- Podemos ver que entre mas temporadas sacan, el IMDB disminuye
# 3.- Las demas son correlaciones bajas
```

```
# GRAFICA PARA CORROBORAR RELACIONES ENTRE VARIABLES 
x = df.IMDB
y = df['Episode ']
z = df['Watch Time (minutes)']
plt.plot(x, y, '*')
plt.plot(x, z, '*')
plt.title('Grafico de acuerdo al IMDB y Watch Time')
plt.xlabel('IMDB')
plt.ylabel('Episodio   Duracion')
plt.legend(['Episodios', 'Duracion']) 
plt.show()

```
